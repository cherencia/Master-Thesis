```python
# import relevant packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the data
data = pd.read_csv(r'campaign clicksa.csv')
data.head()
```

```python
data.describe()
```

```python
data['CTR'] = data['CAMPAIGN_CLICKS']/data['CAMPAIGN_IMPRESSIONS']
data
```

```python
data.describe()
```

```python
data.plot(x='date', 
             y='CTR', 
             title='CTR')
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.lineplot(data['date'], data['CTR'] )


ax.set_xlabel('Date', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_ylabel('CTR', fontsize = 16, fontdict=dict(weight='bold'))
ax.set_xticks(ax.get_xticks()[::15])
plt.tick_params(axis='y', which='major', labelsize=16)
plt.tick_params(axis='x', which='major', labelsize=16)
```

```python
# matplotlib histogram
plt.hist(data['CTR'], color = 'blue', edgecolor = 'black',
         bins = int(1/0.05))

# seaborn histogram
sns.distplot(data['CTR'], hist=True, kde=False, 
             bins=int(1/0.05), color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Histogram of CTR')
plt.xlabel('CTR')
plt.ylabel('Days')
```

```python
# Density Plot and Histogram of all arrival delays
sns.distplot(data['CTR'], hist=True, kde=True, 
             bins=int(1/0.05), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```
